```python
#CODE TO GET THE 3 LARGEST PROGRAMS
import pandas as pd
df = pd.read_excel("output/cbo_projection_changes.xlsx", sheet_name="changes by TIN")
df.shape
```

```python
df['percent_change'] = df['percent_change'].astype(float)
#df = df[df['25-34']>0]
```

```python
df[df['Agency'] =="Small Business Administration"].sort_values(by='percent_change', ascending=False)
```

```python
df = df.groupby('Agency').filter(lambda x: len(x) >= 3) #only those with three or more entries
top_three_rows = df.groupby('Agency').apply(lambda x: x.nlargest(3, 'percent_change')).reset_index(drop=True)
top_three_rows
```

```python
filename = "output/cbo_projection_changes.xlsx"
sheetname = "top three programs by agency"
with pd.ExcelWriter(filename, mode="a", engine="openpyxl") as writer:
    top_three_rows.sort_values(by=['Agency','percent_change'], ascending=False).to_excel(writer, sheet_name=sheetname, index=False)
    
```
