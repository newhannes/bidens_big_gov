```python
#CODE TO GET THE 3 LARGEST INCREASES of whatever variable
#assumes there already exists a column with "dollar_change"
#should "dollar_change" be abs(2024) - abs(2021)? Or just regular values?
#imports and read in data with little cleaning
import pandas as pd
case_studies = ["Department of Agriculture", "Department of Commerce", "Department of Labor", 'Department of the Treasury', "Department of Transportation", "Environmental Protection Agency"]
sheetname = "agency-bureau inc 2"
df = pd.read_excel("../output/cbo_projection_changes.xlsx", sheet_name=sheetname).query('percent_change > 0') #filter out decreases in spending
df['percent_change'] = df['percent_change'].astype(float)
df.head(2)
```

```python
#a quick check to see what the data looks like
df[df['Agency'] =="Department of Transportation"].sort_values(by='percent_change', ascending=False)
```

```python
df = df.groupby('Agency').filter(lambda x: len(x) >= 3) #only those with three or more entries
col_for_largest = "percent_change" 
top_three = df.groupby('Agency').apply(lambda x: x.nlargest(3, col_for_largest)).reset_index(drop=True)
top_three
```

```python
df.query('Agency == "Department of Commerce"').sort_values(by='percent_change', ascending=False)
```

```python
df = top_three.query('Agency in @case_studies')
df['percent_change'] = round(df['percent_change'])
df.to_excel("../output/top_three_agencies.xlsx", sheet_name="3 largest by percent", index=False)
```

```python
filename = "../output/program_increases.xlsx"
sheetname = "top three bureaus for agency"
df_final = top_three.query('Agency in @case_studies')
with pd.ExcelWriter(filename, mode="a", engine="openpyxl") as writer:
    df_final.sort_values(by=['Agency',col_for_largest], ascending=False).to_excel(writer, sheet_name=sheetname, index=False)
```
